import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**2 - 1) / (x - 1)

x = np.linspace(0.5, 2.5, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$y = \frac{x^2 - 1}{x - 1}$")
plt.axvline(x=1, color='red', linestyle='--', label='x = 1')
plt.axhline(y=2, color='green', linestyle='--', label='y = 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de y = (x^2 - 1) / (x - 1)')
plt.legend()
plt.grid(True)
plt.show()
# No gráfico, podemos ver que a função tem um "buraco" em x = 1 porque não está definida nesse ponto. No entanto, à medida que x se aproxima de 1, y se aproxima de 2. Isso é indicado pela linha tracejada horizontal em  y = 2 e pelo comportamento da curva conforme x se aproxima de 1.
